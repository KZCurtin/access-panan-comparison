# 1 month salt budget test for panant-01 control, shelf only

Converting salt budget terms into freshwater fluxes so I can compare the magnitude of these terms to the freshwater volume fluxes.

```python
import glob
import dask.distributed as dsk
import matplotlib.pyplot as plt
import xarray as xr
import cf_xarray as cfxr
import numpy as np
import xesmf as xe
from matplotlib.colors import to_rgba

import cmocean as cm
import cartopy.crs as ccrs

import cartopy.feature as cft
import matplotlib.path as mpath

import warnings # ignore these warnings
warnings.filterwarnings("ignore", category = FutureWarning)
warnings.filterwarnings("ignore", category = UserWarning)
warnings.filterwarnings("ignore", category = RuntimeWarning)
```

```python
lat_slice = slice(-79,-59)

# Importing data
paths_1_C = np.sort(glob.glob('/scratch/x77/kc5856/mom6/archive/panant-01-zstar-ACCESSyr2-test-pan01_test_01-f7e0ea96/output091/*.ocean_month.nc'))
paths_2_C = np.sort(glob.glob('/scratch/x77/kc5856/mom6/archive/panant-01-zstar-ACCESSyr2-test-pan01_test_01-f7e0ea96/output091/*.ocean_month_z.nc'))

paths_1_M = np.sort(glob.glob('/scratch/x77/kc5856/mom6/archive/panant-ssp126-sofia-expt-3393fa55/output091/*.ocean_month.nc'))
paths_2_M = np.sort(glob.glob('/scratch/x77/kc5856/mom6/archive/panant-ssp126-sofia-expt-3393fa55/output091/*.ocean_month_z.nc'))
```

```python
# Time variant variables
def preprocess_1(ds):
    ds = ds[['wfo','lrunoff','salt_flux','salt_flux_added']].sel(yh=lat_slice)
    return ds

def preprocess_2(ds):
    ds = ds[['so','S_advection_xy','Sh_tendency_vert_remap','osalttend','osaltpmdiff','osaltdiff','boundary_forcing_salt_tendency']].sel(yh=lat_slice)
    return ds

data_1_C = xr.open_mfdataset(paths_1_C, preprocess = preprocess_1, chunks = 'auto')
data_2_C = xr.open_mfdataset(paths_2_C, preprocess = preprocess_2, chunks = 'auto')

data_1_M = xr.open_mfdataset(paths_1_M, preprocess = preprocess_1, chunks = 'auto')
data_2_M = xr.open_mfdataset(paths_2_M, preprocess = preprocess_2, chunks = 'auto')
```

```python
# Shelf mask

def shelf_mask_isobath(var, model_dict):

    paths = {
             "mom5": "/g/data/ik11/grids/Antarctic_slope_contour_1000m.npz",
             "mom6_01": "/g/data/ik11/grids/Antarctic_slope_contour_1000m_MOM6_01deg.nc",
            "mom6_005": "/g/data/ik11/grids/Antarctic_slope_contour_1000m_MOM6_005deg.nc"
             }

    var = var.cf.sel({'latitude': slice(-90, -59)})

    if paths[model_dict][-3:] == '.nc':
        shelf_mask = xr.open_dataset(paths[model_dict])['contour_masked_above']
    else:
        contour_file = np.load(paths[model_dict])
        shelf_mask = xr.DataArray(contour_file['contour_masked_above'],
                                  coords = var.coords, 
                                  dims = var.dims,
                                  name = 'contour_masked_above')
    
    shelf_mask = xr.where(shelf_mask == 0, 1, 0)
    masked_var = var * shelf_mask
    
    return masked_var, shelf_mask
```

```python
depth = xr.open_dataset('/g/data/ol01/outputs/mom6-panan/panant-01-zstar-ACCESSyr2/output050/20050501.ocean_static.nc')['deptho']
land_mask = (0 * depth).fillna(1)
land = xr.where(np.isnan(depth.rename('land_1')), 1, np.nan)
depth_shelf, shelf_mask = shelf_mask_isobath(depth, 'mom6_01')
```

```python
# testing individual diagnostics
tendency_C = data_2_C.osalttend
adv_horizontal_C = data_2_C.S_advection_xy
adv_vertical_C = data_2_C.Sh_tendency_vert_remap
surface_forcing_C = data_2_C.boundary_forcing_salt_tendency
diff_1_C = data_2_C.osaltdiff
diff_2_C = data_2_C.osaltpmdiff
SF_C = data_1_C.salt_flux
SFA_C = data_1_C.salt_flux_added

# testing individual diagnostics
tendency_M = data_2_M.osalttend
adv_horizontal_M = data_2_M.S_advection_xy
adv_vertical_M = data_2_M.Sh_tendency_vert_remap
surface_forcing_M = data_2_M.boundary_forcing_salt_tendency
diff_1_M = data_2_M.osaltdiff
diff_2_M = data_2_M.osaltpmdiff
SF_M = data_1_M.salt_flux
SFA_M = data_1_M.salt_flux_added
```

```python
# Converting diagnostics to freshwater fluxes
tendency_FWF_M = -tendency_M*1000/SSS_M  # negative sign makes this a salt -> FW flux conversion
adv_horizontal_FWF_M = -adv_horizontal_M*1000/SSS_M
adv_vertical_FWF_M = -adv_vertical_M*1000/SSS_M
surface_forcing_FWF_M = -surface_forcing_M*1000/SSS_M
diff_1_FWF_M = -diff_1_M*1000/SSS_M
diff_2_FWF_M = -diff_2_M*1000/SSS_M
SF_FWF_M = -SF_M*1000/SSS_M
SFA_FWF_M = -SFA_M*1000/SSS_M
```

```python
# Load files
pan = np.load("/g/data/x77/kc5856/analysis/access-panan-comparison/salt_freshwater_terms_pan01.npz")
acc = np.load("/g/data/x77/kc5856/analysis/access-panan-comparison/salt_freshwater_terms_access.npz")

# Extract base names (remove _C and _M)
pan_terms = set([k[:-2] for k in pan.files])
acc_terms = set([k[:-2] for k in acc.files])

# Union of all terms
all_terms = sorted(pan_terms.union(acc_terms))
```

```python
# Terms to combine: key = new name, value = list of terms to sum
combine_terms = {
    "Total Advection": ["adv_hor", "advection"],
    "Melt\n(ice+snow)": ["melt","net_melt"],
    "Total FWF": ["pme_river","wfo"],
    "Runoff": ["river","lrunoff"],
    "Surface salt - restoring": ["surface","vert_diff_surface - pme_correct"],
    "Salt tendency": ["tendency"],
    "Pan01 restoring": ["salt_flux_added"],
    "ACCESS cont restoring": ["salt_restoring"],
    "ACCESS melt restoring": ["pme_correct"],
}

# Terms to exclude
exclude_terms = [
    "wfiform","wfimelt","budget","vert_diff_surface","adv_vert",
    "diff1","eta_smooth","river_mix","evap","fprec","nonlocal_KPP",
    "submeso","precip","pme_net"
]

# Terms to prioritize for ordering (so they appear next to each other)
special_order = ["ACCESS cont restoring", "ACCESS melt restoring"]
```

```python
pan_anom = []
acc_anom = []
plot_labels = []

# Start with all_terms from the union
all_terms = sorted(set([k[:-2] for k in pan.files]).union(set([k[:-2] for k in acc.files])))

# Keep track of which terms we've already combined
used_terms = set()

for term in all_terms:
    if term in used_terms or term in exclude_terms:
        continue  # skip already combined or excluded terms

    # Check if this term should be combined
    combined = False
    for new_name, term_list in combine_terms.items():
        if term in term_list:
            pan_sum = 0.0
            acc_sum = 0.0
            for t in term_list:
                if t in exclude_terms:
                    continue
                used_terms.add(t)
                pan_val = pan[f"{t}_M"] - pan[f"{t}_C"] if f"{t}_M" in pan.files and f"{t}_C" in pan.files else 0.0
                acc_val = acc[f"{t}_M"] - acc[f"{t}_C"] if f"{t}_M" in acc.files and f"{t}_C" in acc.files else 0.0
                pan_sum += pan_val
                acc_sum += acc_val
            if not (pan_sum == 0 and acc_sum == 0):
                pan_anom.append(pan_sum)
                acc_anom.append(acc_sum)
                plot_labels.append(new_name)
            combined = True
            break
    if combined:
        continue

    # Normal term
    pan_val = pan[f"{term}_M"] - pan[f"{term}_C"] if f"{term}_M" in pan.files and f"{term}_C" in pan.files else 0.0
    acc_val = acc[f"{term}_M"] - acc[f"{term}_C"] if f"{term}_M" in acc.files and f"{term}_C" in acc.files else 0.0

    if pan_val == 0 and acc_val == 0:
        continue

    pan_anom.append(pan_val)
    acc_anom.append(acc_val)
    plot_labels.append(term)

# Reorder special terms to be next to each other
for t in reversed(special_order):
    if t in plot_labels:
        idx = plot_labels.index(t)
        # move to end temporarily
        plot_labels.append(plot_labels.pop(idx))
        pan_anom.append(pan_anom.pop(idx))
        acc_anom.append(acc_anom.pop(idx))
```

```python
# Plotting
x = np.arange(len(plot_labels))
width = 0.4

fig, ax = plt.subplots(figsize=(12,6))

ax.bar(x - width/2, pan_anom, width, color='black', label='pan01')
ax.bar(x + width/2, acc_anom, width, color='red', label='access')

ax.axhline(0, color='black', linewidth=1)

ax.set_xticks(x)
ax.set_xticklabels(plot_labels, rotation=45, ha='right')

ax.set_ylabel("Freshwater flux anomaly (kg m-2 s-1)")
ax.set_title("Budget Term Anomalies: pan01 vs access")
ax.legend(loc='lower left')

plt.tight_layout()
plt.show()
```
