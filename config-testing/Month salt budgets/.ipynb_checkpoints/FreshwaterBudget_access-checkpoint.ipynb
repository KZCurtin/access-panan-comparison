```python
lat_slice = slice(-79,-59)

# Importing data
paths_2_C = np.sort(glob.glob('/scratch/x77/kc5856/access-om2/archive/01deg_jra55v13_ryf9091_21mbath_copy-15025726/output040/ocean/ocean.nc'))
paths_1_C = np.sort(glob.glob('/scratch/x77/kc5856/access-om2/archive/01deg_jra55v13_ryf9091_21mbath_copy-15025726/output040/ocean/ocean_month.nc'))

paths_2_M = np.sort(glob.glob('/scratch/x77/kc5856/access-om2/archive/access-ssp126-sofia-expt-c085baf0/output040/ocean/ocean.nc'))
paths_1_M = np.sort(glob.glob('/scratch/x77/kc5856/access-om2/archive/access-ssp126-sofia-expt-c085baf0/output040/ocean/ocean_month.nc'))
```

```python
# Time variant variables
def preprocess_1(ds):
    ds = ds[['pme_river','pme_net','pme_correct','river','wfimelt','wfiform','fprec','lprec','evap','runoff']].sel(yt_ocean=lat_slice)
    return ds

def preprocess_2(ds):
    ds = ds[['salt']].sel(yt_ocean=lat_slice)
    return ds

data_1_C = xr.open_mfdataset(paths_1_C, preprocess = preprocess_1, chunks = 'auto')
data_2_C = xr.open_mfdataset(paths_2_C, preprocess = preprocess_2, chunks = 'auto')

data_1_M = xr.open_mfdataset(paths_1_M, preprocess = preprocess_1, chunks = 'auto')
data_2_M = xr.open_mfdataset(paths_2_M, preprocess = preprocess_2, chunks = 'auto')
```

```python
# Sea surface salinity - maybe not needed
SSS_C = data_2_C.salt.isel(st_ocean=0)
```

```python
# Doing spatial average
area = xr.open_dataset('/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/output012/ocean/ocean_grid.nc')['area_t']
area_corr = area*(SSS_C*0 +1)
area_corr.plot()
```

```python
# Shelf mask

def shelf_mask_isobath(var, model_dict):

    paths = {
             "mom5": "/g/data/ik11/grids/Antarctic_slope_contour_1000m.npz",
             "mom6_01": "/g/data/ik11/grids/Antarctic_slope_contour_1000m_MOM6_01deg.nc",
            "mom6_005": "/g/data/ik11/grids/Antarctic_slope_contour_1000m_MOM6_005deg.nc"
             }

    var = var.cf.sel({'latitude': slice(-90, -59)})

    if paths[model_dict][-3:] == '.nc':
        shelf_mask = xr.open_dataset(paths[model_dict])['contour_masked_above']
    else:
        contour_file = np.load(paths[model_dict])
        shelf_mask = xr.DataArray(contour_file['contour_masked_above'],
                                  coords = var.coords, 
                                  dims = var.dims,
                                  name = 'contour_masked_above')
    
    shelf_mask = xr.where(shelf_mask == 0, 1, 0)
    masked_var = var * shelf_mask
    
    return masked_var, shelf_mask
```

```python
depth = xr.open_dataset('/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/output012/ocean/ocean_grid.nc')['ht']
land_mask = (0 * depth).fillna(1)
land = xr.where(np.isnan(depth.rename('land_1')), 1, np.nan)
depth_shelf, shelf_mask = shelf_mask_isobath(depth, 'mom5')
```

```python
# Shelf only 
pme_net_C = data_1_C.pme_net.where(shelf_mask == 1)
pme_river_C = data_1_C.pme_river.where(shelf_mask == 1)
river_C = data_1_C.river.where(shelf_mask == 1)
wfimelt_C = data_1_C.wfimelt.where(shelf_mask == 1)
wfiform_C = data_1_C.wfiform.where(shelf_mask == 1)
melt_C = (wfiform_C + wfimelt_C)

pme_net_M = data_1_M.pme_net.where(shelf_mask == 1)
pme_river_M = data_1_M.pme_river.where(shelf_mask == 1)
river_M = data_1_M.river.where(shelf_mask == 1)
wfimelt_M = data_1_M.wfimelt.where(shelf_mask == 1)
wfiform_M = data_1_M.wfiform.where(shelf_mask == 1)
melt_M = (wfiform_M + wfimelt_M)
pme_correct_M = data_1_M.pme_correct.where(shelf_mask == 1)
```

```python
shelf_area = area_corr.where(shelf_mask == 1).sum(['xt_ocean', 'yt_ocean'])

# Spatial average
pme_net_ave_C = (area_corr * pme_net_C).sum(['xt_ocean', 'yt_ocean']) / shelf_area
pme_river_ave_C = (area_corr * pme_river_C).sum(['xt_ocean', 'yt_ocean']) / shelf_area
river_ave_C = (area_corr * river_C).sum(['xt_ocean', 'yt_ocean']) / shelf_area
wfimelt_ave_C = (area_corr * wfimelt_C).sum(['xt_ocean', 'yt_ocean']) / shelf_area
wfiform_ave_C = (area_corr * wfiform_C).sum(['xt_ocean', 'yt_ocean']) / shelf_area
melt_ave_C = (area_corr * melt_C).sum(['xt_ocean', 'yt_ocean']) / shelf_area

pme_net_ave_M = (area_corr * pme_net_M).sum(['xt_ocean', 'yt_ocean']) / shelf_area
pme_river_ave_M = (area_corr * pme_river_M).sum(['xt_ocean', 'yt_ocean']) / shelf_area
river_ave_M = (area_corr * river_M).sum(['xt_ocean', 'yt_ocean']) / shelf_area
wfimelt_ave_M = (area_corr * wfimelt_M).sum(['xt_ocean', 'yt_ocean']) / shelf_area
wfiform_ave_M = (area_corr * wfiform_M).sum(['xt_ocean', 'yt_ocean']) / shelf_area
melt_ave_M = (area_corr * melt_M).sum(['xt_ocean', 'yt_ocean']) / shelf_area
pme_correct_ave_M = (area_corr * pme_correct_M).sum(['xt_ocean', 'yt_ocean']) / shelf_area 
```

```python
pme_net_ave_C = pme_net_ave_C.squeeze().load()
pme_river_ave_C = pme_river_ave_C.squeeze().load()
river_ave_C = river_ave_C.squeeze().load()
wfimelt_ave_C = wfimelt_ave_C.squeeze().load()
wfiform_ave_C = wfiform_ave_C.squeeze().load()
melt_ave_C = melt_ave_C.squeeze().load()

pme_net_ave_M = pme_net_ave_M.squeeze().load()
pme_river_ave_M = pme_river_ave_M.squeeze().load()
river_ave_M = river_ave_M.squeeze().load()
wfimelt_ave_M = wfimelt_ave_M.squeeze().load()
wfiform_ave_M = wfiform_ave_M.squeeze().load()
melt_ave_M = melt_ave_M.squeeze().load()
pme_correct_ave_M = pme_correct_ave_M.squeeze().load()
```

The surface FW budget is given by:  
PME_RIVER = RIVER + MELT (wfiform+wfimelt) + PME (pme_net - (wfiform+wfimelt)) 

```python
# Term labels
terms = [
    "pme_river",
    "pme_net",
    "river",
    "wfimelt",
    "wfiform",
    "melt",
    "pme_correct",
    "budget",
]

# Anomaly values
control = [
    pme_river_ave_C.values,
    pme_net_ave_C.values,
    river_ave_C.values,
    wfimelt_ave_C.values,
    wfiform_ave_C.values,
    melt_ave_C.values,
    0,
    (river_ave_C + melt_ave_C + (pme_net_ave_C - melt_ave_C)).values
]

# Anomaly values
melt = [
    pme_river_ave_M.values,
    pme_net_ave_M.values,
    river_ave_M.values,
    wfimelt_ave_M.values,
    wfiform_ave_M.values,
    melt_ave_M.values,
    pme_correct_ave_M.values,
    (river_ave_M + melt_ave_M + (pme_net_ave_M - melt_ave_M) - pme_correct_ave_M).values
]
```

I need to understand why we need to subtract `pme_correct_ave_M` in the budget. Is it maybe already included on the RHS of the budget but not in the LHS (`pme_river`). So we have really just got `pme_river` + `pme_correct` = old RHS as the budget equation now?

```python
# Bar positions
x = np.arange(len(terms))
width = 0.38

# Create figure
plt.figure(figsize=(10, 5))
plt.bar(x - width/2, control, width, label='Control')
plt.bar(x + width/2, melt, width, label='Melt')

plt.axhline(0, color='black', linewidth=0.7)
plt.ylabel("Freshwater flux (kg m-2 s-1)")
plt.xticks(x, terms, rotation=45)
plt.title("Depth-integrated Shelf-Averaged Budget Terms")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Checking closure: Control
budget_C = river_ave_C + melt_ave_C + (pme_net_ave_C - melt_ave_C)
residual_C = pme_river_ave_C - budget_C

# Checking closure: Melt
budget_M = river_ave_M + melt_ave_M + (pme_net_ave_M - melt_ave_M) - pme_correct_ave_M
residual_M = pme_river_ave_M - budget_M

print("Control")
print(f"  pme_river_ave_C = {float(pme_river_ave_C.values):.6e}")
print(f"  budget_C        = {float(budget_C.values):.6e}")
print(f"  residual_C      = {float(residual_C.values):.6e}")

print("\nMelt")
print(f"  pme_river_ave_M = {float(pme_river_ave_M.values):.6e}")
print(f"  budget_M        = {float(budget_M.values):.6e}")
print(f"  residual_M      = {float(residual_M.values):.6e}")
```
