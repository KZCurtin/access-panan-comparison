# Checking meltwater input

```python
import matplotlib.pyplot as plt
import xarray as xr
import cf_xarray as cfxr
import numpy as np
import glob

import cartopy.feature as cft
import matplotlib.path as mpath
import cmocean as cm

import warnings # ignore these warnings
warnings.filterwarnings("ignore", category = FutureWarning)
warnings.filterwarnings("ignore", category = UserWarning)
warnings.filterwarnings("ignore", category = RuntimeWarning)
```

```python
lat_slice = slice(-79,-59)

# Importing data
pathsC_1 = np.sort(glob.glob('/g/data/ol01/outputs/mom6-panan/panant-01-zstar-ACCESSyr2/*/*.ocean_month.nc'))[24:]
pathsM_1 = np.sort(glob.glob('/scratch/x77/jn8053/mom6/archive/panant-01-zstar-ssp126-prsc-rest/*/*.ocean_month.nc'))

pathsC_2 = np.sort(glob.glob('/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/outpu*/ocean/ocean_month.nc'))
pathsM_2 = np.sort(glob.glob('/scratch/x77/jn8053/access-om2/archive/01deg_jra55v13_ryf9091_21mbath_sofia_ssp126_prsc_rest/*/ocean/ocean_month.nc'))
```

```python
# Time variant variables
def preprocess_1(ds):
    ds = ds[['wfo','lrunoff']].sel(yh=lat_slice)
    return ds

cont_pan = xr.open_mfdataset(pathsC_1, preprocess = preprocess_1, chunks = 'auto')
melt_pan = xr.open_mfdataset(pathsM_1, preprocess = preprocess_1, chunks = 'auto')
```

```python
# Time variant variables
def preprocess_2(ds):
    ds = ds[['pme_river','runoff']].sel(yt_ocean=lat_slice)
    return ds

cont_acc = xr.open_mfdataset(pathsC_2, preprocess = preprocess_2, chunks = 'auto')
melt_acc = xr.open_mfdataset(pathsM_2, preprocess = preprocess_2, chunks = 'auto')
```

First just doing a visual comparison.

```python
# Taking time mean
wfo_pan_C = cont_pan['wfo'].sel(time=slice('2001-01','2010-12')).mean(dim='time')
wfo_pan_M = melt_pan['wfo'].sel(time=slice('2001-01','2010-12')).mean(dim='time')

wfo_acc_C = cont_acc['pme_river'].sel(time=slice('1910-01','1919-12')).mean(dim='time')
wfo_acc_M = melt_acc['pme_river'].sel(time=slice('1910-01','1919-12')).mean(dim='time')

# Anomaly
wfo_pan_anom = wfo_pan_M - wfo_pan_C
wfo_acc_anom = wfo_acc_M - wfo_acc_C
```

```python
plt.figure(figsize=(18, 5))
wfo_pan_anom.plot(vmin=-0.0004,vmax=0.0004,cmap = cm.cm.balance)
plt.ylim(-80,-65)
```

```python
plt.figure(figsize=(18, 5))
wfo_acc_anom.plot(vmin=-0.0004,vmax=0.0004,cmap = cm.cm.balance)
plt.ylim(-80,-65)
```

Now doing it with runoff specifically

```python
# Taking time mean
run_pan_C = cont_pan['lrunoff'].sel(time=slice('2001-01','2010-12')).mean(dim='time')
run_pan_M = melt_pan['lrunoff'].sel(time=slice('2001-01','2010-12')).mean(dim='time')

run_acc_C = cont_acc['runoff'].sel(time=slice('1910-01','1919-12')).mean(dim='time')
run_acc_M = melt_acc['runoff'].sel(time=slice('1910-01','1919-12')).mean(dim='time')

# Anomaly
run_pan_anom = run_pan_M - run_pan_C
run_acc_anom = run_acc_M - run_acc_C
```

```python
plt.figure(figsize=(18, 5))
run_pan_anom.plot(vmin=-0.0004,vmax=0.0004,cmap = cm.cm.balance)
plt.ylim(-80,-65)
```

```python
plt.figure(figsize=(18, 5))
run_acc_anom.plot(vmin=-0.0004,vmax=0.0004,cmap = cm.cm.balance)
plt.ylim(-80,-65)
```

```python
# Testing just one timestep
run_pan_test_C = cont_pan['lrunoff'].sel(time=slice('2001-01','2010-12')).mean(dim='time')
run_pan_test_C = run_pan_test_C.where(run_pan_test_C != 0)
run_pan_test_M = melt_pan['lrunoff'].sel(time=slice('2001-01','2010-12')).mean(dim='time')
run_pan_test_M = run_pan_test_M.where(run_pan_test_M != 0)

run_acc_test_C = cont_acc['runoff'].sel(time=slice('1910-01','1919-12')).mean(dim='time')
run_acc_test_C = run_acc_test_C.where(run_acc_test_C != 0)
run_acc_test_M = melt_acc['runoff'].sel(time=slice('1910-01','1919-12')).mean(dim='time')
run_acc_test_M = run_acc_test_M.where(run_acc_test_M != 0)

# Anomaly
run_pan_test_anom = run_pan_test_M - run_pan_test_C
run_acc_test_anom = run_acc_test_M - run_acc_test_C
```
